```python
# Import relevant libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Connect to and load data
url = 'https://raw.githubusercontent.com/IbrahimHammed/MachineLearning/refs/heads/main/Advertising_2023.csv'
df = pd.read_csv(url)
```

```python
# Review top 5 records of the dataframe
df.head()
```

```python
# Start feature engineering
selectedFeatures = ['digital', 'TV', 'radio', 'newspaper', 'sales']
df = df[selectedFeatures]
df.head()
```

```python
# Generate and rveiew correlation matrix for the dataframe
df.corr()
```

```python
# Plot correlation matrix as heatmap using Seaborn
sns.heatmap(df.corr(), annot=True)
```

```python
# Plot a scatter diagram featuring digital and sales
%matplotlib inline
plt.xlabel('digital')
plt.ylabel('sales')
plt.scatter(df.digital,df.sales,color='red',marker='+')
```

```python
# Plot a scatter diagram featuring TV and sales
%matplotlib inline
plt.xlabel('TV')
plt.ylabel('sales')
plt.scatter(df.TV,df.sales,color='red',marker='+')
```

```python
# Plot a scatter diagram featuring radio and sales
%matplotlib inline
plt.xlabel('radio')
plt.ylabel('sales')
plt.scatter(df.radio,df.sales,color='red',marker='+')
```

```python
# Plot a scatter diagram featuring newspaper and sales
%matplotlib inline
plt.xlabel('newspaper')
plt.ylabel('sales')
plt.scatter(df.newspaper,df.sales,color='red',marker='+')
```

```python
# extract predictors by dropping 'sales' column from the dataframe
x = df.drop('sales', axis='columns')
x.head()
```

```python
# Extract target column 'sales'
y = df['sales']
y.head()
```

```python
# Create linear regression model
regressionmodel = linear_model.LinearRegression()
```

```python
# Train the model
regressionmodel.fit(x, y)
```

```python
# Make Prediction
regressionmodel.predict([[345.15, 156, 37.8, 69.2]])
```

```python
# Coefficients
regressionmodel.coef_
```

```python
# Intercept
regressionmodel.intercept_
```

```python
# Evaluate model
regressionmodel.score(x, y)
```

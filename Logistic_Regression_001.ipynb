ANN for predicting whether a patient has Kyphosis or not

```python
# Import relevant libraries
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
url = 'https://raw.githubusercontent.com/IbrahimHammed/MachineLearning/refs/heads/main/kyphosis%20(1).csv'
df = pd.read_csv(url)
```

```python
print("\nDataset shape:", df.shape)
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

```python
df.info()
```

```python
df.head(3)
```

```python
# Extracting predictors and output features
x = df.drop(columns=['Kyphosis'])
y = df['Kyphosis']
```

```python
# Train Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
# Feature Scaling
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Model Production
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Model Training
model.fit(x_train, y_train)
```

```python
# Handling predictions
predictions = model.predict(x_test)
```

```python
# Evaluate model
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test, predictions))
```

```python
# Accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
